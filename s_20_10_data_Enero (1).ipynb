```python
print(dataset.columns)
```

```python
# Assuming 'dataset' is your DataFrame as defined in the previous code.
unique_sizes = dataset['TAMAÑO '].unique()
unique_sizes
```

```python
# Convertir la columna 'TAMAÑO ' a numérica, manejando errores
dataset['TAMAÑO '] = pd.to_numeric(dataset['TAMAÑO '], errors='coerce')

# Ahora, realiza la agrupación y agregación
dataset.groupby('TUMOR PRIMARIO')['TAMAÑO '].aggregate(['min','max','mean']) \
        .rename(columns={'min':'Minimo', 'max':'Maximo', 'mean':'Promedio'}) \
        .sort_values(by='Maximo', ascending=False)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(16,6))
```

```python
fig = dataset.groupby('TUMOR PRIMARIO')['TAMAÑO '].mean().plot.bar(color='green')
fig.set_title('Tamaño Promedio del Tumor Primario', {'fontsize':22})
fig.set_ylabel('Tamaño Promedio del Tumor Primario ')
```

```python
import numpy as np  # Importa numpy con el alias 'np'

boxplot = dataset.boxplot(column='TAMAÑO ', by='TUMOR PRIMARIO', vert=False)
boxplot.set_xticks(np.arange(0, 70, 10)) # Ahora 'np' está definido y se puede usar
plt.gca().xaxis.get_major_formatter().set_scientific(False)
```
